```python
plt.plot(df['Date'],df['Open'])
plt.xlabel("Date")
plt.ylabel("Opening Price")
plt.show()
```

```python
plt.plot(df["Date"],df["Close"])
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df=df.sort_values(by="Date")
plt.plot(df["Date"],df["Close"],color="black",label="Closing price v/s Date")
plt.title("Line Plot")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
plt.bar(df["Date"][0:5],df["Open"][0:5],color="black")
plt.title("Date v/s Open")
plt.xlabel("Date")
plt.ylabel("open")
plt.xticks(rotation=45)

plt.show()
```

```python
w=0.6
plt.bar(df["High"][0:5],df["Low"][0:5],w,color="black")
plt.bar(df["High"][0:5],df["Low"][0:5],w,color="gray",bottom=df["High"][0:5])

plt.title("Stacked Bar Plot")
plt.show()
```

```python
fig,ax=plt.subplots()
ax.stackplot(df["Open"],df["High"],df["Low"],df["Close"],labels=["Open","High","Low"],colors=["black","brown","gray"])
plt.legend(loc="upper left")
plt.show()
```
