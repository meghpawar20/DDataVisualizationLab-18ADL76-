```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
np.random.seed(2368575)
npoints= 1000
nbins=20

x=np.random.randn(npoints)
y=.8**x + np.random.randn(1000)+25

fig,axs=plt.subplots(1,1,
                    figsize=(8,4),
                    tight_layout=True)
axs.hist(x,bins=nbins,color="black")
plt.show()
```

```python
fig=plt.figure(figsize=(8,4))
plt.boxplot(y)
plt.show()
```

```python
plt.scatter(x,y,color="gray")
plt.plot(np.unique(x),np.poly1d(np.polyfit(x,y,1))
        (np.unique(x)),color="black")
```

```python
nr=4
nc=2
ni=nr*nc
c=["pink","purple","cyan","orange","blue","red","black","brown"]
plt.figure(figsize=(2*2*nc,2*nr))
for i in range(ni):
    plt.subplot(nr,nc,i+1)
    plt.plot(x,y,color=c[i])
plt.tight_layout()
plt.show()
```
